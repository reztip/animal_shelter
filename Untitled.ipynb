Let's start off with some exploratory data analysis.

How many dogs are there? <b>15,595 in our training set.</b>

What is the relative mix of cat and dogs? <b> About 58% dogs, 42% cats.</b>

What is the outcome distribution, by animal type (see the plot below.)

```python
x = df.AnimalType.value_counts()
print('Number of Animals')
print(x)
print('Proportion of Animals')
print(df.AnimalType.value_counts(normalize = True))
```

```python
sns.countplot(df.AnimalType, hue = df.AnimalType)
```

The <code>seaborn</code> library can make some pretty graphs.  Below is the outcome of each animal by type.

```python
sns.countplot(data = df, x = "OutcomeType", hue = 'AnimalType')
plt.title("Outcome by Animal Type", fontsize = 16, fontweight = 'bold')
```

I can't really see the distribution of deaths, let's take a better look.

```python
sns.countplot(df[df.Died == 1]['AnimalType'].sort_values(ascending = False))
plt.title("Animals Who Died by Type", fontweight = 'bold', fontsize = 16);
```

The idea came to me to see if it appears that neutering status has an effect on shelter outcome.  In the below plot, it appears that neutering is a pretty strong indicator for an animal to be adopted or returned to the owner.  Not being neutered makes euthanasia and animal transfers seem more likely.

```python
sns.countplot(data = df, x = 'OutcomeType', hue = 'Neutered')
plt.title("Outcome by Neuter Status", fontsize = 16, fontweight = 'bold');
```

It turns out most cats are domestic short, medium or long hair mixes. What happened to them?

```python
sns.countplot(data = df[df.Breed.isin(['Domestic Shorthair Mix', 'Domestic Medium Hair Mix', 'Domestic Longhair Mix'])], x = "OutcomeType", hue = "Breed")
plt.title("Outcome For {X}-Hair Mix Cats", fontsize = 16, fontweight = 'bold');
```
